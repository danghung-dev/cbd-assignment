```python
import pandas as pd
import numpy as np # linear algebra
import matplotlib.pyplot as plt
import scipy
import seaborn as sns
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
data = pd.read_csv("./Melbourne_housing_FULL.csv")
data.head()
```

```python

data['Type'].value_counts().head(10).plot.bar()
data['Suburb'].value_counts().head(10).plot.bar()
data['Method'].value_counts().head(10).plot.bar()
```

```python
data['Suburb'].value_counts().head(10).plot.bar()
```

```python
data['Rooms'].value_counts().sort_index().plot.bar(
    figsize = (10,5),
    fontsize = 14,
    title = "Count of Rooms"
)
```

```python

# Multivariant 
sns.lmplot(x='Distance', 
           y='Price',
          hue = 'Type',
          data = data)
```

```python
data_temp = data[data['Rooms'] < 15]
sns.lmplot(x='Rooms', 
           y='Price',
          hue = 'Type',
          data = data_temp)
```

```python
data = data[data.Price.notnull()]


sns.distplot(data['Price'])
```

```python
#Tranform categorical data into numerical data for training purpose
from pandas.api.types import is_string_dtype
for col, col_data in data.items():
    if is_string_dtype(col_data):
        data[col] = data[col].astype('category').cat.as_ordered().cat.codes
```
