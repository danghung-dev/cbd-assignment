# Datasets transformation

```python
def transformAttrition(value):
    if (value == 'Yes'):
        return 1
    if (value == 'No'):
        return 0
data['AttritionNum'] = data.Attrition.apply(lambda x: transformAttrition(x))
data.head()
```

We transformed Attrition string to new numeric column: AttritionNum

# What are key factors that are playing into current attrition rates?


We seperate two group Attrition = YES and Attrtion = NO
We loop all numeric columns. For each column we set hypothesis H0 does not affect Attrition and H1 does affect Attrition rate.
Then we use t test to reject H0 hypothesis.

```python
attrition0 = data[data['AttritionNum'] == 0]
attrition1 = data[data['AttritionNum'] == 1]
affectAttrition = []
for column in data:
    if (data[column].dtype == 'int64'):
      if (column == 'AttritionNum'):
        continue
      print('\n====================%s=========================' % column)
      print('H0: %s does not affect Attrition rate' % column)
      print('H1: %s does affect Attrition rate' % column)
      twosample = scipy.stats.ttest_ind(attrition0[column], attrition1[column])
      print(twosample)
      if (twosample[1] < 0.05):
        print('%s < 0.05, we can reject H0, %s does affect Attrition rate' % (twosample[1], column))
        affectAttrition.append(column)
      else:
        print('%s > 0.05, we cannot reject H0' % twosample[1])

print('\n\n')        
print('==================What are key factors that are playing into current attrition rates?=======================')
print('Key factors affect to Attrition rate', affectAttrition)
```

# What are key factors that are playing into current satisfaction rates?

We seperate 4 group JobSatisfaction = 1,2,3,4
We loop all numeric columns. For each column we set hypothesis H0 does not affect JobSatisfaction and H1 does affect JobSatisfaction rate.
Then we use anova to reject H0 hypothesis.

```python
jobSf1 = data[data['JobSatisfaction'] == 1]
jobSf2 = data[data['JobSatisfaction'] == 2]
jobSf3 = data[data['JobSatisfaction'] == 3]
jobSf4 = data[data['JobSatisfaction'] == 4]
affectJobSf = []
for column in data:
    if (data[column].dtype == 'int64'):
      if (column == 'JobSatisfaction'):
        continue
      print('\n====================%s=========================' % column)
      print('H0: %s does not affect JobSatisfaction rate' % column)
      print('H1: %s does affect JobSatisfaction rate' % column)
      jobAnovaOneWay = scipy.stats.f_oneway(jobSf1[column], jobSf2[column], jobSf3[column], jobSf4[column])
      print(jobAnovaOneWay)
      if (jobAnovaOneWay[1] < 0.05):
        print('%s < 0.05, we can reject H0, %s does affect Attrition rate' % (jobAnovaOneWay[1], column))
        affectJobSf.append(column)
      else:
        print('%s > 0.05, we cannot reject H0' % jobAnovaOneWay[1])

print('\n\n')        
print('==================What are key factors that are playing into current satisfaction rates?=======================')
print(affectJobSf)
```

# When are employees leaving?

Those factors that affect to Attrition rate.

['Age', 'DailyRate', 'DistanceFromHome', 'EnvironmentSatisfaction', 'JobInvolvement', 'JobLevel', 'JobSatisfaction', 'MonthlyIncome', 'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsWithCurrManager']

We will look into those columns and find when employees leaving.

```python
dfgb = data.groupby(['JobSatisfaction','EnvironmentSatisfaction']).agg({'Attrition': [ pd.Series.count, pd.Series.nunique],
                                                 'AttritionNum': pd.Series.sum})
dfgb['AttritionRate'] = dfgb['AttritionNum']['sum'] / dfgb['Attrition']['count']
dfgb
```

Conclusion: If JobSatisfaction and EnvironmentSatisfaction are low, Attrition rate will be high.

```python
dfgb = data.groupby(['JobInvolvement','JobLevel']).agg({'Attrition': [ pd.Series.count, pd.Series.nunique],
                                                 'AttritionNum': pd.Series.sum})
dfgb['AttritionRate'] = dfgb['AttritionNum']['sum'] / dfgb['Attrition']['count']
dfgb
```
