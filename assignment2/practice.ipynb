```python
import pandas as pd
import scipy
data = pd.read_csv("./data.csv")
data.head()

```

```python
def transformAttrition(value):
    if (value == 'Yes'):
        return 1
    if (value == 'No'):
        return 0
data['AttritionNum'] = data.Attrition.apply(lambda x: transformAttrition(x))
data.head()
```

```python
attrition0 = data[data['AttritionNum'] == 0]
attrition1 = data[data['AttritionNum'] == 1]
```

```python
import matplotlib.pyplot as plt
import numpy as np
bins_age = np.arange(start=15, stop=65, step=5)
AgeAttritionArrange1 = np.histogram(attrition1['Age'], bins=bins_age)
AgeAttritionArragee0 = np.histogram(attrition0['Age'], bins=bins_age)
AgeAttritionRate = AgeAttritionArrange1[0]/AgeAttritionArragee0[0]

print(AgeAttritionRate)
plt.hist(data['Age'], bins=bins_age)
plt.title('Histogram of Age Att1')
plt.show()
```
